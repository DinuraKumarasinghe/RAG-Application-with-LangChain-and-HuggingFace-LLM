### Initialize Embedding Model

```python
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
  model_name="sentence-transformers/all-mpnet-base-v2"
)
```

### Initialize Output Parser

```python
from langchain.schema.output_parser import StrOutputParser

output_parser=StrOutputParser()
```

```python
from langchain_community.document_loaders import PyPDFLoader

# Load the PDF document
loader = PyPDFLoader("/content/Profile.pdf")

docs = loader.load()
```

### Split Documents into Chunks

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Initialize the text splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=400, chunk_overlap=50)

# Split the documents into chunks
splits = text_splitter.split_documents(docs)
```

```python
len(splits)
```

### Create Vector Store and Retriever

```python
from langchain_chroma import Chroma

# Create a vector store from the document chunks
vectorstore = Chroma.from_documents(documents=splits, embedding=embedding_model)
```

```python
# Create a retriever from the vector store
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

### Define Prompt Template

```python
from langchain.prompts import ChatPromptTemplate

# Define prompt template
template = """
Answer this question using the provided context.

{question}

Context:
{context}

Answer:
"""

prompt=ChatPromptTemplate.from_template(template)
```

```python
prompt
```

```python
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

# Create a prompt template with a dynamic conversation history and a new question placeholder
prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="Answer this question using the provided context "),
        MessagesPlaceholder(variable_name="history"),
        MessagesPlaceholder(variable_name="question")
    ]
)

# Chain the prompt, LLM, and output parser
chain = prompt | llm | output_parser
```

```python
# Define the conversation history
history = [
    HumanMessage(content="My name is Dinura"),
    AIMessage(content="Nice to meet you, Dinura! How can I assist you today?"),
    HumanMessage(content="what is 2 + 2"),
    AIMessage(content="4")
]
```

```python
question = "what was the univercity Dinura Kumarasinghe went to"

response = chain.invoke({"history": history, "question": [HumanMessage(content=question)]})

print(response)
```

```python
# Extend the history with the latest question and response
history.extend([HumanMessage(content=question), AIMessage(content=response)])
```

### Chain Retriever and Prompt Template with LLM

```python
from langchain.schema.runnable import RunnablePassthrough

chain = (
    {"context": retriever,  "question": RunnablePassthrough()}
    | prompt
    | llm
    | output_parser
)
```

#### Invoke RAG Chain with Example Questions

```python
response = chain.invoke("who is Dinura Kumarasinghe?")
print(response)
```

```python
response = chain.invoke(" what was the univercity Dinura Kumarasinghe went to")
print(response)
```
